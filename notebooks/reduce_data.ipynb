```python
# # Update the df to reduce the number of data points for each unique "dataset"
# # to be 1/16 of the original number of data points.
df["datasetkey"] = df["dataset"].astype("category").cat.codes
df_new = (
    df.groupby("datasetkey")
    .apply(lambda x: x.sample(frac=1 / 6), include_groups=False)
    .reset_index(drop=True)
)
df_new

# import numpy as np

# # We already have the sampled indices -- just use those
# idxs = np.load("./small_idxs.npy")
# df_new = df.loc[idxs].reset_index(drop=True)
# df_new
```

```python
from pymatgen.core.structure import Molecule


def update(structure: Molecule):
    if not isinstance(structure, Molecule):
        return structure

    # Update positions to center around the origin
    coords = structure.cart_coords
    coords -= coords.mean(axis=0)
    structure = Molecule(structure.species, coords)

    return structure


df_new["structure"] = df_new["structure"].apply(update)
```
