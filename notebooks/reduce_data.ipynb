```python
from pymatgen.core import Molecule, Structure
from typing_extensions import assert_never


def update(structure: Molecule | Structure):
    match structure:
        case Molecule():
            # Update positions to center around the origin
            coords = structure.cart_coords
            coords -= coords.mean(axis=0)
            structure = Molecule(structure.species, coords)
            # structure.translate_sites(range(len(structure)), -structure.center_of_mass)
        case Structure():
            if True:
                return structure
            # Calculate the geometric center in fractional coordinates
            # Summing up all fractional coordinates and dividing by the number of sites
            total_frac_coords = [0.0, 0.0, 0.0]
            for site in structure:
                frac_coords = site.frac_coords
                total_frac_coords = [
                    x + y for x, y in zip(total_frac_coords, frac_coords)
                ]

            num_atoms = len(structure)
            geometric_center = [x / num_atoms for x in total_frac_coords]

            # Compute the shift needed to move the center to (0.5, 0.5, 0.5) which is the center of the unit cell
            shift_vector = [0.5 - x for x in geometric_center]

            # Apply the shift to all atoms in the structure, ensuring the movement is in fractional coordinates
            structure.translate_sites(range(num_atoms), shift_vector, frac_coords=True)

            # Check and wrap sites to keep within the unit cell
            structure.perturb(
                0
            )  # This small perturbation can help in wrapping sites properly

        case _:
            assert_never(structure)
    return structure


df_new["structure"] = df_new["structure"].apply(update)
```
