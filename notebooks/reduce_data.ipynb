```python
import pandas as pd

df = pd.read_pickle("/workspaces/fm/fm/notebooks/webpage/df.pkl")
df[df["dataset"] == "qm9"]
```

```python
# Update the df to reduce the number of data points for each unique "dataset"
# to be 1/16 of the original number of data points.
df["datasetkey"] = df["dataset"].astype("category").cat.codes
df_new = (
    df.groupby("datasetkey")
    .apply(lambda x: x.sample(frac=1 / 16), include_groups=False)
    .reset_index(drop=True)
)
df_new
```

```python
df_new.to_pickle("../df_small.pkl")
```
