```python
import time
from pathlib import Path

import nglview
from pymatgen.core import Structure

row = df.iloc[0]
display(row)

structure: Structure = row["structure"]
display(structure)


def save_image(view: nglview.NGLWidget, output_path: Path):
    img = view.render_image()

    i = 0
    while not img.value:
        if i % 10 == 0:
            print("Waiting for image to render...")
        time.sleep(0.1)
        i += 1

    with output_path.open("wb") as f:
        f.write(img.value)


def render_structure_to_file(structure: Structure, dest: Path):
    # Import the necessary modules
    import nglview

    # Create a view object
    view = nglview.show_pymatgen(structure)

    # Render the view to a file
    save_image(view, dest)


render_structure_to_file(structure, Path("./structure.png"))
# # Display the image
# from IPython.display import Image

# Image(filename="./structure.png")
```
